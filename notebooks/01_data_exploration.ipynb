```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv(r"C:\Users\satya\mycotoxin\mycotoxin_prediction\data\MLE-Assignment.csv")

# Check Data Types
print(df.dtypes)

# Handling Missing Values - Only Numeric Columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Visualization: Distribution of Vomitoxin (DON) Concentration
plt.figure(figsize=(8, 5))
sns.histplot(df['vomitoxin_ppb'], bins=30, kde=True)
plt.title("Distribution of Vomitoxin (DON) Concentration")
plt.xlabel("Vomitoxin Concentration (ppb)")
plt.ylabel("Frequency")
plt.show()

# Line plot of average spectral reflectance over wavelengths
plt.figure(figsize=(10, 5))
plt.plot(df.iloc[:, 1:].mean(), color='b')
plt.xlabel("Wavelength Index")
plt.ylabel("Mean Reflectance")
plt.title("Average Spectral Reflectance Over Wavelengths")
plt.show()

# Save cleaned dataset
df.to_csv(r"C:\Users\satya\mycotoxin\mycotoxin_prediction\data\cleaned_dataset.csv", index=False)
```
